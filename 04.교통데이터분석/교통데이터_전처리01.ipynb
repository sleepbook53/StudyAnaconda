# 컬럼명 한글명으로 변환

##### 1. 영문명과 한글명이 있는 메타정의서 조회
##### 2. 딕셔너리로 가져오기
##### 3. rename(딕셔너리 넣기) : 컬럼명 변경시에 딕셔너리 형태로 넣어야 사용 가능

```python
# 메타 정의서 조회/ 영문명과 한글명이 있는 컬럼만 조회

df_bus_card_col_org = pd.read_excel("./01_data_원본/포항시_버스_사용자_교통카드_사용_내역/trfcard (0)/trfcard_columns.xlsx"
                                   , header = 2
                                   , usecols = "B:C")
df_bus_card_col_org
```

```python
#딕셔너리로 변환 작업 1: .to_dict("recrods")  - 리스트형태의 딕셔너리
df_bus_card_col_org_dict = df_bus_card_col_org.to_dict("records")
#records: 데이터 프레임의 리스트 행 의미 -> 각 행들을 딕셔너리로 전환할 것
```

```python
#딕셔너리로 변환작업2 - 완전한 딕셔너리 만들기
#key와 value를 직접 만들기

#깨끗한 삽입용 데이터 베이스 생성
df_bus_card_new_dict = {}

#for 문
for i in range(0, len(df_bus_card_col_org_dict)) :
    df_bus_card_new_dict[df_bus_card_col_org_dict[i]["컬럼명 (영문)"]]= df_bus_card_col_org_dict[i]["컬럼명 (한글)"]
    #df_bus_card_new_dict["key"] = "value"
    #df_bus_card_new_dict["key2"] = "value2"
    #{key : value, key2 : value2}

```

```python
#컬럼명 변경
df_bus_card_kor=df_bus_card_org.rename(columns = df_bus_card_new_dict)
```

# 분석 주제 정하기 : 분석요소(컬럼) 확인

```python
# [버스 이용량 분석]
# 기준일 및 기준일별 버스 이용량 분석 비교
# 기준일 및 시간대별 번스 이용량 분석 비교
# 기준 시간 및 시간(분)별 버스 이용량 분석 비교

# [버스 내 체류 시간]
# [승차정류장별 버스내 체류시간]
```

```python
df_bus_card_kor.info()
```

```python
# 승차시각 및 하차시각에 대해서 월/일/시간/분을 추출 하기 위해
#문자타입으로 변경

df_bus_card_kor= df_bus_card_kor.astype({"승차시각": "str"})
df_bus_card_kor= df_bus_card_kor.astype({"하차시각": "str"})
```

```python
df_bus_card_kor.info()
```

```python
#분석요소(컬럼)을 추출하여 저장할 데이터 프레임생성
df_bus_card = pd.DataFrame()

df_bus_card["승차시각"] = pd.to_datetime(df_bus_card_kor["승차시각"])
df_bus_card["하차시각"] = pd.to_datetime(df_bus_card_kor["하차시각"])
df_bus_card["승객구분"] = df_bus_card_kor["승객연령"]
df_bus_card["환승여부"] = df_bus_card_kor["환승여부"]
df_bus_card["추가운임여부"] = df_bus_card_kor["추가운임여부"]
df_bus_card["승차정류장"] = df_bus_card_kor["승차정류장"]
df_bus_card["하차정류장"] = df_bus_card_kor["하차정류장"]
```

```python
df_bus_card
```

```python
#버스내 체류간(분) 계산 컬럼 만들기
#버스내 체류시간(분) = 하차시각(분) -  승차시각(분)
df_bus_card["버스내체류시각(분)"] = \
            pd.to_datetime(df_bus_card["하차시각"]).dt.minute - \
            pd.to_datetime(df_bus_card["승차시각"]).dt.minute

#양식이 date라서 다음과 같이 해도 가능
#df_bus_card["하차시각"].dt.minute - \
            df_bus_card["승차시각"].dt.minute
```

# 모든 데이터셋 통합하기

```python
from datetime import datetime
```

```python
#실행시간 체크하기 => 시작시간
start_date = datetime.today().strftime("%y-%m-%d-%H:%M:%S")

#여기에 통합할 프로그램 작성
df_bus_card_tot = pd.DataFrame()

for i in range(0, 100, 1) :
    # 1. csv, excel 파일 불러오기
    file_path1 = ("./01_data_원본/포항시_버스_사용자_교통카드_사용_내역/trfcard ({})/trfcard.csv").format(i)
    file_path2 = ("./01_data_원본/포항시_버스_사용자_교통카드_사용_내역/trfcard ({})/trfcard_columns.xlsx").format(i)
                                
    df_bus_card_org     = pd.read_csv(file_path1)
    df_bus_card_col_org = pd.read_excel(file_path2,  header = 2, usecols = "B:C")
    
    # 2. 파일명 한글명 바꾸기
    df_bus_card_col_dict = df_bus_card_col_org.to_dict("records")
    
    df_bus_card_col_new_dict = {}
    for i in range(0, len(df_bus_card_col_dict), 1) :
        df_bus_card_col_new_dict[df_bus_card_col_dict[i]["컬럼명 (영문)"]] = \
                                df_bus_card_col_dict[i]["컬럼명 (한글)"]
         
        df_bus_card_kor = df_bus_card_org.rename(columns = df_bus_card_col_new_dict)
    # 3. 분석요소 추출 및 추가하기
    df_bus_card_kor = df_bus_card_kor.astype({"승차시각" : "str"} )
    df_bus_card_kor = df_bus_card_kor.astype({"하차시각" : "str"} )
    
    df_bus_card = pd.DataFrame()
    
    df_bus_card["승차시각"]     = pd.to_datetime(df_bus_card_kor["승차시각"])
    df_bus_card["하차시각"]     = pd.to_datetime(df_bus_card_kor["하차시각"])
    df_bus_card["승객구분"]     = df_bus_card_kor["승객연령"]
    df_bus_card["환승여부"]     = df_bus_card_kor["환승여부"]
    df_bus_card["추가운임여부"] = df_bus_card_kor["추가운임여부"]
    df_bus_card["승차정류장"]   = df_bus_card_kor["승차정류장"]
    df_bus_card["하차정류장"]   = df_bus_card_kor["하차정류장"]
    
    df_bus_card["버스체류시간(분)"] = \
                df_bus_card["하차시각"].dt.minute - \
                df_bus_card["승차시각"].dt.minute
    df_bus_card.loc[(df_bus_card["버스체류시간(분)"]<0),"버스체류시간(분)"] =\
                    60- (df_bus_card["버스체류시간(분)"] * -1)
    
    df_bus_card["기준월"] = df_bus_card["승차시각"].dt.month
    df_bus_card["기준일"] = df_bus_card["승차시각"].dt.day
    df_bus_card["기준시간"] = df_bus_card["승차시각"].dt.hour
    df_bus_card["기준분"] = df_bus_card["승차시각"].dt.minute
    
    # 4. 데이터 추가(병합) 하기
    df_bus_card_tot = df_bus_card_tot.append(df_bus_card, ignore_index = True)

end_date = datetime.today().strftime("%y-%m-%d-%H:%M:%S")
print("전체 실행 시간 : (", start_date, ") ~(", end_date,")")
print("전체 데이터 갯수 : ", len(df_bus_card_tot))
df_bus_card_tot.head()
```

```python
#파일저장
df_bus_card_tot.to_csv("./02_data_통합본/df_bus_card_tot.csv", index=None)
```
