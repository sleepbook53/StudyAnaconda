# 컬럼명 한글명으로 변환
### 샘플 trfcard(0)폴더에 있는 파일을 기준으로 우선변환

```python
import pandas as pd
```

```python
sample = pd.read_csv("./01_data_원본/포항시_버스_사용자_교통카드_사용_내역/trfcard (0)/trfcard.csv")
```

```python
#컬럼 정의서 가져오기
sample_m = pd.read_excel("./01_data_원본/포항시_버스_사용자_교통카드_사용_내역/trfcard (0)/trfcard_columns.xlsx",
                              sheet_name = "컬럼 정의서"
                              , usecols ="B, C"
                              , header = 2)
```

```python
sample_m
```

```python
#한글명으로 칼럼명 변경

#데이터 값들을 딕셔너리로 변경
sample_list= sample_m.to_dict("records")

#깨끗한 삽입용 데이터 베이스 생성
new_dict = {}

#ㄹfor 문
for dict_kv in sample_list :
    new_dict[dict_kv["컬럼명 (영문)"]] = dict_kv["컬럼명 (한글)"]

new_dict

sample_kor = sample.rename(columns =  new_dict)
sample_kor
```
