### 최대값/ 최소값 계산하여 이상치 제거

```python
#넘파이 라이브러리 불러들여오기
import numpy as np
```

```python
#q1, q3값 확인
#np.percentil(): 25 %와 75% 시점의 데이터값 알려줌(return)
q1, q3 =np.percentile(df_drop_allrow["나이"],[25,75])
print(q1)
print(q3)
```

```python
#iqr 값 개산하기
iqr = q3-q1
iqr
```

```python
#최댓값/최소값 계산하기 
#최대값 = q3 + (iqr* 1.5)
#최소값 = q1 - (iqr * 1.5)

upper_bound = q3 + (iqr *1.5)
upper_bound #61.5
```

```python
#최소값 계산하기
lower_bound = q1 - (iqr*1.5)
lower_bound #1.5
```

### 이상치 데이터 조회

```python
#나이 데이터에서 이상치 값 조회
df_drop_allrow[(df_drop_allrow["나이"] < lower_bound) | (df_drop_allrow["나이"]> upper_bound)]
```

```python
#데이터 갯수
len(df_drop_allrow[(df_drop_allrow["나이"] < lower_bound) | (df_drop_allrow["나이"]> upper_bound)])
```

### 함수

```python
#매개변수가 없는 이상치 조회를 함수화 시키기

def outliers_iqr() :
    #q1, q3값 확인
    #np.percentil(): 25 %와 75% 시점의 데이터값 알려줌(return)
    q1, q3 =np.percentile(df_drop_allrow["나이"],[25,75])
    print(q1)
    print(q3)
    
    #iqr 값 계산하기
    iqr = q3-q1
    print(iqr)
    
    #최솟값/최댓값
    upper_bound = q3 + (iqr *1.5)
    lower_bound = q1 - (iqr*1.5)
    
    #나이 데이터에서 이상치 값 조회
    df_temp = df_drop_allrow[(df_drop_allrow["나이"] < lower_bound) | (df_drop_allrow["나이"]> upper_bound)]
    
    print(len(df_temp))
    print(df_temp)
    
    
```

```python
#변수출력
outliers_iqr()
```

```python
#데이터를 넘겨주는데로 알아서 이상치 값을 출력하는 기능만들
#매개변수가 이쓴 함수 만들기

def outliers_iqr2(data) :
    
    #q1, q3값 확인
    #np.percentil(): 25 %와 75% 시점의 데이터값 알려줌(return)
    q1, q3 =np.percentile(data,[25,75])
    print(q1)
    print(q3)
    
    #iqr 값 계산하기
    iqr = q3-q1
    print(iqr)
    
    #최솟값/최댓값
    upper_bound = q3 + (iqr *1.5)
    lower_bound = q1 - (iqr*1.5)
    
    #나이 데이터에서 이상치 값 조회
    df_temp = df_drop_allrow[(data< lower_bound)
                             | (data > upper_bound)]
    
    print(len(df_temp))
    print(df_temp)
```

```python
 #함수호출
outliers_iqr2(df_drop_allrow["나이"])
```

```python
#매개 변수 2개 함수 만들기
def outliers_iqr3(dframe,data) :
    
    #q1, q3값 확인
    #np.percentil(): 25 %와 75% 시점의 데이터값 알려줌(return)
    q1, q3 =np.percentile(data,[25,75])
    print(q1)
    print(q3)
    
    #iqr 값 계산하기
    iqr = q3-q1
    print(iqr)
    
    #최솟값/최댓값
    upper_bound = q3 + (iqr *1.5)
    lower_bound = q1 - (iqr*1.5)
    
    #나이 데이터에서 이상치 값 조회
    df_temp =dframe[(data< lower_bound)
                             | (data > upper_bound)]
    
    print(len(df_temp))
    print(df_temp)
```

```python
#호출하기
outliers_iqr3(df_drop_allrow, df_drop_allrow["나이"])
```

```python
#이상치가 있는 
def get_outliers_data(data) :
    
    #q1, q3값 확인
    #np.percentil(): 25 %와 75% 시점의 데이터값 알려줌(return)
    q1, q3 =np.percentile(data,[25,75])
    print(q1)
    print(q3)
    
    #iqr 값 계산하기
    iqr = q3-q1
    print(iqr)
    
    #최솟값/최댓값
    upper_bound = q3 + (iqr *1.5)
    lower_bound = q1 - (iqr*1.5)
    
    #나이 데이터에서 이상치 값 조회
   # df_temp =dframe[(data< lower_bound)
                             #| (data > upper_bound)]
    
   # print(len(df_temp))
    #print(df_temp)
    return np.where((data > 61.5) | (data < 1.5))
```

```python
get_outliers_data(df_drop_allrow["나이"])
```

```python
#전체 데이터프레임에서 행번호로 찾기
df_drop_allrow.loc[get_outliers_data(df_drop_allrow["나이"])]
```

### 이상치 제거하기

```python
#이상치가 아닌 값만 추출, 따로 저장
age_non_outlier_index = []

#이상치 없는 정상 인덱스만 추출
for idx in df_drop_allrow.index : 
    if idx not in age_outlier_index[0] : #0번째 컬럼: 인덱스칸 지정
        age_non_outlier_index.append(idx)

#append() 리스트 추가

age_non_outlier_index
```

```python
age_outlier_index[0]
```

```python
#전체 데이터프레임에서 정상데이터만 조회
df_new = df_drop_allrow.loc[age_non_outlier_index]
df_new 
```

```python
#인덱스 번호 재배열후 최종결과물 변수에 저장 후 조회
df_new = df_new.reset_index(drop=True)
df_new
```
