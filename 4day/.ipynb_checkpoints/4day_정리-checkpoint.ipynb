# 함수

## 매개변수 없는 이상치 조회를 함수화 시키기

```python
def outliers_iqr() :
    #q1, q3값 확인
    #np.percentil(): 25 %와 75% 시점의 데이터값 알려줌(return)
    q1, q3 =np.percentile(df_drop_allrow["나이"],[25,75])
    print(q1)
    print(q3)
    
    #iqr 값 계산하기
    iqr = q3-q1
    print(iqr)
    
    #최솟값/최댓값
    upper_bound = q3 + (iqr *1.5)
    lower_bound = q1 - (iqr*1.5)
    
    #나이 데이터에서 이상치 값 조회
    df_temp = df_drop_allrow[(df_drop_allrow["나이"] < lower_bound) | (df_drop_allrow["나이"]> upper_bound)]
    
    print(len(df_temp))
    print(df_temp)
    
```

```python
#변수출력
outliers_iqr()
```

## 매개변수가 있는 함수 만들기

### 데이터를 넘겨주는대로 알아서 이상치 값을출력하는 기능

```python
def outliers_iqr2(data) :
    
    #q1, q3값 확인
    #np.percentil(): 25 %와 75% 시점의 데이터값 알려줌(return)
    q1, q3 =np.percentile(data,[25,75])
    print(q1)
    print(q3)
    
    #iqr 값 계산하기
    iqr = q3-q1
    print(iqr)
    
    #최솟값/최댓값
    upper_bound = q3 + (iqr *1.5)
    lower_bound = q1 - (iqr*1.5)
    
    #나이 데이터에서 이상치 값 조회
    df_temp = df_drop_allrow[(data< lower_bound)
                             | (data > upper_bound)]
    
    print(len(df_temp))
    print(df_temp)
```

```python
 #함수호출
outliers_iqr2(df_drop_allrow["나이"])
```

## 매개변수 2개가 있는 함수 만들기

```python
def outliers_iqr3(dframe,data) :
    
    #q1, q3값 확인
    #np.percentil(): 25 %와 75% 시점의 데이터값 알려줌(return)
    q1, q3 =np.percentile(data,[25,75])
    print(q1)
    print(q3)
    
    #iqr 값 계산하기
    iqr = q3-q1
    print(iqr)
    
    #최솟값/최댓값
    upper_bound = q3 + (iqr *1.5)
    lower_bound = q1 - (iqr*1.5)
    
    #나이 데이터에서 이상치 값 조회
    df_temp =dframe[(data< lower_bound)
                             | (data > upper_bound)]
    
    print(len(df_temp))
    print(df_temp)
```

```python
#호출하기
outliers_iqr3(df_drop_allrow, df_drop_allrow["나이"])
```

## 이상치 있는 데이터만 출력

```python
def get_outliers_data(data) :
    
    #q1, q3값 확인
    #np.percentil(): 25 %와 75% 시점의 데이터값 알려줌(return)
    q1, q3 =np.percentile(data,[25,75])
    print(q1)
    print(q3)
    
    #iqr 값 계산하기
    iqr = q3-q1
    print(iqr)
    
    #최솟값/최댓값
    upper_bound = q3 + (iqr *1.5)
    lower_bound = q1 - (iqr*1.5)

    return np.where((data > upper_bound) | (data < lower_bound))
```

```python
get_outliers_data(df_drop_allrow["나이"])
```

#### 전체 데이터프레임에서 행번호로 찾기
df_drop_allrow.loc[get_outliers_data(df_drop_allrow["나이"])]

```python
# 함수값 변수로 받아오기
age_outlier_index = get_outliers_data(df_drop_allrow["나이"])
```

#### 확인

##### df_drop_allrow.loc[age_outlier_index] 행의 우치만 확인해서 조회

# 이상치 제거하기

```python
#이상치가 아닌 값만 추출, 따로 저장
age_non_outlier_index = []

#이상치 없는 정상 인덱스만 추출
#전체 데이터 프레임 인덱스 번호 읽어 들이기

for idx in df_drop_allrow.index : 
    
    #이상치 인덱스가 아닌 정상 인덱스만 추출
    if idx not in age_outlier_index[0] : #0번째 컬럼: 인덱스칸 지정
        
        #정상 인덱스 값을 리스트에 추가
        age_non_outlier_index.append(idx)

        
#append() 리스트 추가

age_non_outlier_index
```

```python
df_new = df_drop_allrow.loc[age_non_outlier_index]
df_new 
```
