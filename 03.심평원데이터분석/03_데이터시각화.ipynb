# 데이터 시각화 분석

## 월별, 연련댕=별 본인부담 비용 시각화 분석

```python
import pandas as pd
```

```python
df_month_age_any = pd.read_excel("./files_new/df_month_age_any_월별연령대별본인부담비용.xlsx")
df_month_age_any
```

```python
import matplotlib.pyplot as plt
import seaborn
```

```python
pivot_month_age =df_month_age_any.pivot_table(values = "수진자연령"
                                             , index = "요양개시월"
                                             , columns = "연령대"
                                             ,aggfunc ="count")
pivot_month_age["90대"] = pivot_month_age["90대"].fillna(0)
pivot_month_age
```

## 선그래프
