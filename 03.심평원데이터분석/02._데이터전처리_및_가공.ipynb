```python
#20TABLE  한글컬럼명으로 변경된 파일 읽어 들이기
import pandas as pd
df_20TABLE_org = pd.read_excel("./files_new/df_20TABLE_kor_명세서일반내역.xlsx")

df_20TABLE_org
```

```python
#수진자연령, 요양게시일자, 심결본인부담금
df_20TABLE_org[["수진자연령", "요양개시일자", "심결본인부담금"]]
```

```python
#데이터프레임 새로 만들기
#데이터프레임 이름 df_month_age
df_month_age = df_20TABLE_org[["수진자연령", "요양개시일자", "심결본인부담금"]]
df_month_age.head(1)
```

```python
#결측치 확인 - 정보확인
df_month_age.info()
```

```python
#이상치 확인 - 통계치
\df_month_age.describe()
```

```python
#연령대 구간 : 9세이하, 10대, 20대, 90대, 100세이상
df_9under= df_month_age[df_month_age["수진자연령"]<= 9].reset_index(drop = True)
df_9under["연령대"] = "9세이하"

df_10= df_month_age[(df_month_age["수진자연령"]>=10) & (df_month_age["수진자연령"] <=19)].reset_index(drop = True)
df_10["연령대"] = "10대"


df_20= df_month_age[(df_month_age["수진자연령"]>=20) & (df_month_age["수진자연령"] <=29)].reset_index(drop = True)
df_20["연령대"] = "20대"


df_30= df_month_age[(df_month_age["수진자연령"]>=30) & (df_month_age["수진자연령"] <=39)].reset_index(drop = True)
df_30["연령대"] = "30대"

df_40= df_month_age[(df_month_age["수진자연령"]>=40) & (df_month_age["수진자연령"] <=49)].reset_index(drop = True)
df_40["연령대"] = "40대"

df_50= df_month_age[(df_month_age["수진자연령"]>=50) & (df_month_age["수진자연령"] <=59)].reset_index(drop = True)
df_50["연령대"] = "50대"

df_60= df_month_age[(df_month_age["수진자연령"]>=60) & (df_month_age["수진자연령"] <=69)].reset_index(drop = True)
df_60["연령대"] = "60대"

df_70= df_month_age[(df_month_age["수진자연령"]>=70) & (df_month_age["수진자연령"] <=79)].reset_index(drop = True)
df_70["연령대"] = "70대"

df_80= df_month_age[(df_month_age["수진자연령"]>=80) & (df_month_age["수진자연령"] <=89)].reset_index(drop = True)
df_80["연령대"] = "80대"

df_90= df_month_age[(df_month_age["수진자연령"]>=90) & (df_month_age["수진자연령"] <=99)].reset_index(drop = True)
df_90["연령대"] = "90대"
                          
```

```python
#모든 DF합치기
df_age =pd.concat([df_9under
                   , df_10
                   , df_20
                   , df_30
                   , df_40
                   , df_50
                   , df_60
                   , df_70
                   , df_80
                   , df_90]
                  , axis = 0)
df_age.reset_index(drop =True)
```
