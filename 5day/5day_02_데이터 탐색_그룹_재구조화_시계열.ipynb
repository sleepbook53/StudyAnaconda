## 2. 재구조화

### pd.cut
: 동일한 길이의 구간으로 나눠서 표현하는 것(범위)

```python
#cut()
df1=pd.cut(df["math"],3)
#각각의 인덱스별 math의 점수가 어떤 구간에 속해있느지 보여줌
df1
```

```python
df2 = df["math"].groupby(df1)
```

```python
df2.head()
```

```python
#df2에 대한 기초 통계(agg() 사용)
df3 = df2.agg(["count", "mean", "std", "min", "max"])
df3
```

### qcut()
: 동일한 크기의 구간으로 나눠서 표현하는 것(갯수)

```python
#qcut()
pd.qcut(df["math"],3)
```

```python
df4 = pd.qcut(df["math"], 3, labels = False)
```

```python
df4
#각구간별 학생들의 위치를 확인할 수 있음
```

```python
#cut, qcut으로 범주를 나눴을 때, 해당 범주에 속하는 데이터 조회 방법
#df4.index
#for idx in df.index : 
    print(idx)
```

```python
df5 =  pd.qcut(df["math"], 3, labels=np.arange(3, 0, -1))
df5
```

##### range 와 arage

```python
#첫 번째값: 시작값
#두 번쨰값: -1
#세번째 값 : 중간값, or 감소값
for i in range(0, 3, 1) :
    print(i)

print("-------")

#증가값을 감소시킬 때는 두번째 값은 +1
for j in range(3,0,-1) :
    print(j)
```

```python
for i in np.arange(0, 3, 1) :
    print(i)

print("------")
for j in np.arange(3,0,-1) :
     print(j)
```

```python
df6 = df["math"].groupby(df5)
df6.head()
```

```python
#기초 통계 조회
#count, mean, std,min, max
df6.agg(["count", "mean", "std", "min" ,"max"])
```

```python
#cut/qcut 사용시 해당 범주에 속하는 데이터 조회하는 방법
#해당 구간의 데이터 조회하기 위해
#원본 데이터 프레임에 구간 데이터를 컬럼으로 추가하기
df11 = pd.cut(df["math"], 3, labels = False)
df11
```
