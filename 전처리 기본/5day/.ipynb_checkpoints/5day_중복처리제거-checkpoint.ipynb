### 데이터 가져오기

```python
import numpy as np
import pandas as pd
#bicycle_dup.csv.읽어낸 후 조회

file_path = "./data/bicycle_dup.csv"
df = pd.read_csv(file_path)
```

```python
df
```

### 중복데이터 확인하기

duplicated에 keep을 지정 안하면 중복값 중에 첫번 째 값은 Flase<br>
중복된 값 중 첫번째를 제외하고 모두 Ture<br>
제거시에는 True인 행만 제거

##### df전체 데이터 중 '이용거리'라는 컬럼에 대해서만 중복체크

```python
#한 개 컬럼을 기준으로 중복 체크
df.duplicated(["이용거리"])
```

##### 데이터프레임에 '이용거리' 컬럼에 대해서 중복체크

```python
df['이용거리'].duplicated()
```

### 두개 컬럼 기준으로 중복체크하기

```python
#'이용거리', '나이'
df.duplicated(["이용거리","나이"])
```

```python
#같은 결과물
df[['이용거리','나이']].duplicated()
```

```python
#중복이 있는 데이터 조회
len(df[df.duplicated(["이용거리","나이"])])
```

```python
len(df.loc[df.duplicated(["이용거리", "나이"])])
```

```python
np.where(df.duplicated(["이용거리","나이"]))
```

### 중복데이터  keep = "first"

첫 번째는 살려두고 나머지 다 제거

```python
#keep 생략하면 기본: keep ="first"있다
#가장 많이 사용하는 중복체크 방식
df.duplicated(["이용거리"], keep = 'first')
```

### keep = "last"

나머지 다 제거하고 마지막 것만 살려둠

```python
df.duplicated(["이용거리"], keep = 'last')
```

### Keep = False
모든 중복되어 있는 것들 삭제

```python
#keep = False
df.duplicated(["이용거리"], keep = False)
```

```python
#이용거리, 나이 두개 컬럼을 기준으로 중복되는 모든데이터 조회
len(df[df.duplicated(["이용거리", '나이'], keep = False)])
```

##### 깔끔하기

```python
df_dup =df.duplicated(["이용거리", '나이'], keep = False)
```

```python
len(df[df_dup])
```

## 중복데이터 제거하기

### 첫번째 중복값만 남기고 나머지 중복 데이터 제거하기

제거 후 정상적인데이터만 조회

```python
df.drop_duplicates(["이용거리"], keep = 'first')
```

```python
#마지막 중복값들만 남기고, 나머지 제거하기
df.drop_duplicates(["이용거리"], keep = "last")
```

```python
#중복되는 모든 값 제거하고 조회하기
df.drop_duplicates(["이용거리"], keep = False)
```

```python
df_dup = df.drop_duplicates(["이용거리"], keep = False)
```

```python
#중복검증
len(df_dup[df_dup.duplicated(["이용거리"],keep = False)])
```

```python
#인덱스 정렬
df_dup.reset_index(drop = True)
```

```python
#현재 최종 결과물(데이터프레임)에 대한 기초 통계
df_dup.describe()
```

```python
#중복데이터 확인
df2 = df.duplicated(["이용거리"], keep = "first")
df2
```

```python
df2.describe() 
#전체 2건
#고유값 2건
#중복된 것 뺴면 257건
```

```python
#중복된 값
df[df2]
```

```python
#전체값에 중복된 값을 뺀 갯수
len(df) - 163
```

```python
#빈도분석을 이용해서 확인가능
df2.value_counts()
```
