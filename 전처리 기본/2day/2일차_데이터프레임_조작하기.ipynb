## 데이터 합하기

```python
#시리즈 데이터 합하기
#시리즈변수명 s1 ==> 값은 0,1 / 인덱스 값은 A. B로 시리즈 만들기
#시리즈변수명 s2 ==> 값은 2, 3, 4/ 인덱스 값은 a,b,c로 시리즈 만들기

s1 = pd.Series([0,1], index=["A", "B"])


s2 = pd.Series([2,3,4], index=["a", "b", "c"])

```

## 컬럼 단위로 합치기

```python
df_temp1 = pd.DataFrame([["Dog",3], ["Cat", 10], ["Tiger", 6], ["Moose", 3]],
                        index =[0, 1, 2, 3],
                        columns = ["동물", "나이"])
```

```python
df_temp2 = pd.DataFrame([["집",0], ["초원",1],["수풀", 0], ["초원",1 ]], 
                        index = [0, 1, 2, 3],
                        columns = ["사는곳", "뿔의개수"])
```

```python
pd.concat([df_temp1, df_temp2], axis=1)
```

```python
df_info1 =pd.DataFrame({"고객번호" : [1001, 1002, 1003, 1004, 1005, 1006, 1007],
                      "이름" : ["강", "김", "홍", "이", "박", "최", "소"]},
                      columns = ["고객번호", "이름"])
```

```python
df_info1
```

```python
df_info2 = pd.DataFrame({"고객번호": [1001, 1001, 1005, 1006, 1008, 1001],
                        "금액" : [10000, 20000, 15000, 5000, 10000, 30000]},
                       columns = ["고객번호", "금액"])
df_info2
```

```python
#칼럼 단위로 옆으로 추가하기
#단, 고객 번호가 같은 값들만
pd.merge(df_info1, df_info2)
```

```python
#양쪽 다 있으면 있는대로 없으면 없는대로
pd.merge(df_info1, df_info2, how="outer")
```

```python
#leaf outr join
pd.merge(df_info1, df_info2, how = "left")
```

```python
pd.merge(df_info1, df_info2, how = "right")
```
