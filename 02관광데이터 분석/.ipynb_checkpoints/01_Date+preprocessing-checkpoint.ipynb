### 함수 만들기

```python
def create_kto_data(yy, mm) :
    file_path = "./files/kto_{}{}.xlsx".format(yy,str(mm).zfill(2))
    df = pd.read_excel(file_path, header = 1, skipfooter =4, usecols="A:G")
    # print(len(df))

    #["기준년월"] = "2019-01"
    df["기준년월"] = "{}-{}".format(yy, str(mm).zfill(2))



    #국적의 대륙 값들을 관리하는 목록 만들기
    #- 기준이 없기 대문에 기준 만들기
    continetis_list = ["아시아주", "미주", "구주", "대양주","아프리카주"
              , "기타대륙", "교포소계"]

    #국적데이터에 대륙기준값이 포함되어있는지 확인
    #포함되어 있으면 True, 없으면 False가 조회
    condition= df["국적"].isin(continetis_list)
    df_country = df[condition == False]

    #인덱스 정렬
    df_country_newindex = df_country.reset_index(drop =True)

    #아시아(25), 아프리카(5), 유럽(23), 오세아니아(3), 아프리카(2), 기타대륙, 교포
    continents = ["아시아"] * 25 + ["아메리카"]*5 + ["유럽"] *23 + ["오세아니아"] * 3 + ["아프리카"] *2 + ["기타대륙"] + ["교포"]
    df_country_newindex["대륙"] = continents

    #국가별 내 관광객 비율을 계산하여
    #'관광객 비율(%)'이름으로 컬럼 만들기
    #값은 비율 계산하여 넣기
    #비율값, 소수점 한 자리

    df_country_newindex["관광객 비율(%)"] = round(df_country_newindex["관광"]
                                             /df_country_newindex["계"] * 100 ,1)
    df_country_newindex["관광객 비율(%)"]
    #전체 관광객 대비 나라의 관광객수 비율
    #컬럼명 전체 비율(%)
    #값 : 각국가 관광객수/ 전세계 관광객수의 합 * 100
    #소수점 한자리
    df_country_newindex["전체 비율(%)"] = round(df_country_newindex["관광"]
                                        / sum(df_country_newindex["관광"])*100,1)
    df_country_newindex["전체 비율(%)"]

    return df_country_newindex

   
```

```python
#함수 만들기
#함수이름 : create_kto_data(xxx, xxx)
df_all  = pd.DataFrame()
for yy in range(2010, 2020) :
    for mm in range(1, 13) :
        try : 
            temp = create_kto_data(yy, mm)
            df_all= df_all.append(temp, ignore_index=True)
                #print(df_country_newindex)
        except :
            pass
        
len(df_all)
```
