## HEAT MAP

전체 시계열 데이터에 대한 이슈를 한눈에 보기<br>
변수 : 3개가 필요

```python
#년도별 , 월별 조회
#문자열 자르기
df["년도"] = df["기준년월"].str.slice(0,4) 
df["월"] =df["기준년월"].str.slice(5,7)
df.head()
```

```python
df_fitter = df[(df["국적"] == "중국")]
df_fitter
```

### 히트맵 그리는 순서
1. 피봇으로 데이터 집계<br>
2. 피봇 결과를 히트맵에 적용

```python
#피봇으로 데이터 집계
df_pivot = df_fitter.pivot_table(values = "관광"
                     , index = "년도"
                     , columns = "월" )
```

```python
#피봇 결과를 히트맵에 적용
#시본 라이브러리 사용

import seaborn as sns
```

```python
#그래프 그리기
plt.figure(figsize = (16, 10))

#피봇 데이터 넣어서 그래프 그리기
sns.heatmap(df_pivot, annot = True, fmt = ".0f", cmap ="rocket_r")

#타이틀 넣기
plt.title("년/월별 [중국] 관광객 히트뱁 시각화분석" )


#보여주기
plt.show()
```

```python
#국가별로 히트맵 보기

#년도별 , 월별 조회
#문자열 자르기
df["년도"] = df["기준년월"].str.slice(0,4) 
df["월"] =df["기준년월"].str.slice(5,7)
df.head()


cntry_list = ["중국", "일본", "대만", "미국", "홍콩"]

for cntry in cntry_list :
    
    condition = (df["국적"] == cntry)
    condition
    
    df_filter = df[condition]

    
    #피봇으로 데이터 집계
    df_pivot = df_filter.pivot_table(values = "관광"
                     , index = "년도"
                     , columns = "월" )


    #그래프 그리기
    plt.figure(figsize = (16, 10))

    #피봇 데이터 넣어서 그래프 그리기
    sns.heatmap(df_pivot, annot = True, fmt = ".0f", cmap ="rocket_r")

    #타이틀 넣기
    plt.title("년/월별 [{}] 관광객 히트맵 시각화분석".format(cntry))


    #보여주기
    plt.show()
```
